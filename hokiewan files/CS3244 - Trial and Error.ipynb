```python
from hyperas.distributions import choice, uniform
from hyperas import optim
from hyperopt import Trials, STATUS_OK, tpe
```

```python
def data():
    import pandas as pd
    import numpy as np
    from keras.utils import to_categorical
    
    sl_train = pd.read_csv('sign_mnist_train.csv')
    test = pd.read_csv('blackthenshrinkALLDATA.csv')
    
    X_train = np.array(sl_train.iloc[:,1:])
    y_train = to_categorical(np.array(sl_train.iloc[:,0]))

    X_train = X_train.reshape(X_train.shape[0], 28, 28, 1)
    X_train = X_train.astype('float32')
    
    X_test=test.drop(columns=['file name','expected output','expected output num'])
    X_test=np.array(X_test.iloc[:,:])
    
    X_test = X_test.reshape(X_test.shape[0], 28, 28, 1)
    X_test = X_test.astype('float32')
    X_test /= 255
    
    y_test = to_categorical(np.array(test['expected output num'].iloc[:]))
    
    return X_train, y_train, X_test, y_test
```
